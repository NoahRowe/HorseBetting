```python
from helper import *
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

main = pd.read_pickle("../Data/clean_main.df")
```

```python
# A notebook to scale selected features relative to the other horse stats in that race. 

# FEATURES TO MAKE RELATIVE:
# avg speed rating, last speed rating, best surface distance, added weight, horse rating, 
# horse win percent, jockey win percent, trainer win percent, days since last run
```

```python
def makeRelative(var):
    arr = np.zeros(len(main))
    races = np.unique(main["race_id"])
    c = 0
    for race_id in races:
        printProgress(c, len(races), round(len(races) * 0.05))
        c += 1
        
        race = main.loc[main.race_id == race_id]
        raceI = race.index
        
        raceVar = race[var].to_numpy().reshape(-1,1)
        scaler = StandardScaler()
        
        scaledVals = scaler.fit_transform(raceVar)
        
        for i in range(len(race)):
            I = raceI[i]
            arr[I] = scaledVals[i]
            
    main[var+"_rel"] = arr
```
