```python
# ENGINEERING ANALYSIS STEPS:
# CALCULATE Vm AND k FROM PAST RACES
# USE THESE TO SOLVE FOR A PROJECTED FINISH TIME, GIVEN A DISTANCE
#  Do not know how far out the posts are for each track - need to figure this out
```

```python
# Finding Vm and K
def addConstants(main):
    
    Vms = np.zeros(len(main))
    Ks = np.zeros(len(main))
    c = 0 
    for i in range(len(main)):
        printProgress(c, len(main), round(len(main)*0.05))
        c +=1
        race = main.iloc[i, :]
        distance = race['distance']
        surface = race["surface"]
        venue = race["venue"]
        
        sectionalDistances = SECTIONAL_DISTANCES[venue + " "  + str(surface) + " " + str(distance)]
        sectionalCount = len(sectionalDistances) 
        
        times = []
        distances = []
        for j in range(1, sectionalCount+1):
            times.append(sum(times) + race["time"+str(j)])
            distances.append(sum(distances) + sectionalDistances[j-1])
              
        x = np.array([[t,t**2/2] for t in times])
        y = np.array(distances)
        
        R = np.linalg.lstsq(x, y, rcond=None)[0]
    
        Vm = R[0]
        K = R[1]/Vm
        
        Vms[i] = Vm
        Ks[i] = K
        
    main["Vm"] = Vms
    main["K"] = Ks
        
#addConstants(main)
#main.to_pickle("../clean_main.df")
```
