```python
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
from helper import *

main = pd.read_pickle("../Data/clean_main.df")
splitDate = pd.to_datetime("1998-07-19")
data = main[main.date < splitDate]
len(data)
```

```python
ratingDic = {}
winDic = {}
strings = []
venues = np.unique(main["venue"])
for venue in venues:
    vData = data.loc[data.venue==venue]
    surfaces = np.unique(vData.surface)
    
    for surface in surfaces:
        sData = vData.loc[data.surface==surface]      
        distances = np.unique(sData["distance"])
        
        for distance in distances:
            dData = sData.loc[data.distance==distance]
            draws = np.unique(dData.draw)
            
            for draw in draws:
                finalData = dData.loc[dData.draw == draw]
                
                meanRating = np.mean(finalData["speed_rating"])
                winPct = np.mean(finalData["won"])
                
                string = venue + " " + str(surface) + " " + str(distance) + " " + str(draw)
                
                if venue =="ST" and surface == 1 and distance == 1800:
                    winPct = 1/14.
                ratingDic[string] = meanRating
                winDic[string] = winPct
                strings.append(string)
```

```python
np.save("post_win_avg.npy", winDic)
np.save("post_avg.npy", ratingDic)
```

```python
def addBarrierPct(data):
    post, distance, venue, surface = data.draw, data.distance, data.venue, data.surface
    string = venue + " " + str(surface) + " " + str(distance) + " " + str(post)
    if string in strings:
        winPct = winDic[string]
    else:
        winPct = 0
    return winPct

main["draw_win_percent"] = main.apply(addBarrierPct, axis=1)
```
