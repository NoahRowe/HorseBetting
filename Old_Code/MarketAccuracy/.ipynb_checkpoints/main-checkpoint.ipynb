```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
from helper import *

from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

main = pd.read_pickle("../Data/main_no_outliers.df")
knnFeat = ["win_odds"]
knnTarg = "lengths_behind"
```

```python
# Perform KNN on the odds to predict LBW (Lengths behind winner)
k = 20
x = []
y = []
logy = []
knny = []
for training, testing in getCrossVal(main, 10):
    knn = KNeighborsRegressor(k)
    
    X_train = training[knnFeat]
    y_train = training[knnTarg]
    X_test = testing[knnFeat]
    y_test = testing[knnTarg]
    
    knn.fit(X_train, y_train)
    
    knnPreds = knn.predict(X_test)
    
    # Accuracy
    RMSE = mean_squared_error(knnPreds, y_test, squared=False)
    print("KNN Accuracy: {:.3f}".format(RMSE))
    
    # LR on log of knn output
    
    logPreds = [m.log(i+0.0001, 10) for i in knnPreds]
    y.extend(y_test)
    
    x.extend(X_test.values)
    logy.extend(logPreds)
    knny.extend(knnPreds)
    
    # Plot what we have so far x-odds, y-LBW
plt.figure(figsize=(10,8), dpi=100)
plt.plot(x, y, "r.", label="Actual Lenghts Behind")
plt.plot(x, knny, "g.", label="KNN Prediction")
plt.plot(x, logy, "b.", label="Log of KNN Prediction")
plt.legend()
plt.ylabel("LBW")
plt.xlabel("Odds")
    
```
