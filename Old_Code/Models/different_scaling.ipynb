```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from model_helper import *

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

main = pd.read_pickle("scaled_main.df")
```

```python
FEATURES = [#"gear_change", 
            "horse_rating_rel", "last_speed_rating_rel", "average_speed_rating_rel",
            "best_surface_distance_rel", "weight_rating_rel", "rest_rating_rel", "going_rating_rel",
            "jockey_win_percent_rel", "trainer_win_percent_rel", "horse_win_percent_rel",
            "horse_norm_record_rel", "trainer_norm_record_rel", "jockey_norm_record_rel", 
            "trainer_jockey_record_rel", "draw_win_percent_rel",
            "avg_sec1_speed_rating_rel", "last_sec1_speed_rating_rel",
            "avg_last_sec_speed_rating_rel", "last_last_sec_speed_rating_rel",
            "avg_speed_rating_ratio_rel", "last_speed_rating_ratio_rel"
           ]
TARGET = "won"
```

```python
from sklearn.preprocessing import minmax_scale
# min max scaled features

def makeMinMax(var):
    arr = np.zeros(len(main))
    races = np.unique(main["race_id"])
    for race_id in races:
        
        race = main.loc[main.race_id == race_id]
        raceI = race.index
        
        raceVar = race[var].to_numpy().reshape(-1,1)
        
        scaledVals = minmax_scale(raceVar)
        
        for i in range(len(race)):
            I = raceI[i]
            arr[I] = scaledVals[i]
            
    main[var+"_mm"] = arr
c = 1
for var in FEATURES:
    #makeMinMax(var[:-4])
    print("Done", c)
    c += 1
main.to_pickle("scaled_main.df")
```

```python
# robust scaled features
from sklearn.preprocessing import RobustScaler

def makeRobust(var):
    arr = np.zeros(len(main))
    races = np.unique(main["race_id"])
    c = 0
    for race_id in races:
        c += 1
        
        race = main.loc[main.race_id == race_id]
        raceI = race.index
        
        raceVar = race[var].to_numpy().reshape(-1,1)
        scaler = RobustScaler()
        scaledVals = scaler.fit_transform(raceVar)
        
        for i in range(len(race)):
            I = raceI[i]
            arr[I] = scaledVals[i]
            
    main[var+"_rb"] = arr
    
c = 1
for var in FEATURES:
    makeRobust(var[:-4])
    print("Done", c)
    c += 1
    
main.to_pickle("scaled_main.df")
```
