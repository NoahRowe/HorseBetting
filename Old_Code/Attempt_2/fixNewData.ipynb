```python
import pandas as pd
import numpy as np

info = pd.read_csv("horse_info.csv")
record = pd.read_csv("past_record.csv")
record["RaceDate"] = pd.to_datetime(record["RaceDate"])
runs = pd.read_csv("total.csv")
```

```python
# # Add jockey record

jockAvg = np.zeros(len(record))
jockey = np.unique(record["Jockey"])

for i in range(len(jockey)):
    temp = record[record.Jockey == jockey[i]]
    tempI = temp.index
    if (i%10)==0:
        print("{}/{}".format(i, len(jockey)))
    
    for j in range(len(temp)):
        temp_2 = temp.iloc[j,:]
        date = temp_2["RaceDate"]
        
        beforeRaces = temp.loc[temp["RaceDate"]<date]
        
        if beforeRaces.shape[0] != 0:
            jockAvg[tempI[j]] = np.mean(temp["Pla"])

record['jockey_ave_rank'] = jockAvg

trainAvg = np.zeros(len(record))
trainer = np.unique(record["Trainer"])

for i in range(len(trainer)):
    temp = record[record.Trainer == trainer[i]]
    tempI = temp.index
    if (i%10)==0:
        print("{}/{}".format(i, len(trainer)))
    
    for j in range(len(temp)):
        temp_2 = temp.iloc[j,:]
        date = temp_2["RaceDate"]
        
        beforeRaces = temp.loc[temp["RaceDate"]<date]
        
        if beforeRaces.shape[0] != 0:
            trainAvg[tempI[j]] = np.mean(temp["Pla"])

record['trainer_ave_rank'] = trainAvg

recent = np.zeros(len(record))

horses = np.unique(record["HorseID"])
for i in range(len(horses)):
    temp = record[record.HorseID == horses[i]]
    tempI = temp.index
    if (i%10)==0:
        print("{}/{}".format(i, len(horses)))
    
    for j in range(len(temp)):
        temp_2 = temp.iloc[j,:]
        date = temp_2["RaceDate"]
        
        beforeRaces = temp.loc[temp["RaceDate"]<date]
        
        if beforeRaces.shape[0] != 0:
            if beforeRaces.shape[0] <= 6:
                recent[tempI[j]] = np.mean(temp["Pla"])
            else:
                recent[tempI[j]] = np.mean(temp["Pla"][-6:])

record['recent_ave_rank'] = recent

# Add HorseWin
horseWin = np.zeros(len(record), dtype=int)

for i in range(len(record)):
    if record.iloc[i,:]["Pla"]==1:
        horseWin[i] = 1
        
record["HorseWin"] = horseWin
    
```
