```python
# Need to add a number of races
def addNumberOfRaces(data):
    raceNumber = np.zeros(len(data))
    
    horses = np.unique(data["horse_id"])
    for h in horses:
        horseRaces = data.loc[data["horse_id"]==h]
        # Sort it
        horseRaces = horseRaces.sort_values(by=["date"])
        raceSize = horseRaces.shape[0]
        raceI = horseRaces.index
        
        for i in range(raceSize):
            raceNumber[raceI[i]] = i
            
    data["number_of_races"] = raceNumber
    return data

# Add winnings to date
def winningsToDate(data):
    winnings = np.zeros(len(data))
    
    horses = np.unique(data["horse_id"])
    for h in horses:
        horseRaces = data.loc[data["horse_id"]==h]
        horseRaces = horseRaces.sort_values(by=["date"])
        
        raceSize = horseRaces.shape[0]
        raceI = horseRaces.index
        
        for i in range(1, raceSize):
            prevWinnings = winnings[raceI[i-1]]
            wonLastRace = horseRaces.iloc[i-1,:]["won"]
            if wonLastRace==1:
                currentWin = prevWinnings + horseRaces.iloc[i-1,:]["prize"]
            else:
                currentWin = prevWinnings
            winnings[raceI[i]] = currentWin
            
    data["winnings_to_date"] = winnings
    return data

# Add winnings per race
def winningsPerRace(data):
    winnings = np.zeros(len(data))
    
    for i in range(len(data)):
        run = data.iloc[i,:]
        w = run["winnings_to_date"]
        n = run["number_of_races"]
        if n!=0:
            winnings[i] = float(w)/n
            
    data["winnings_per_race"] = winnings
    return data

# Add number of wins
def addNumberOfWins(data):
    wins= np.zeros(len(data))
    
    horses = np.unique(data["horse_id"])
    for h in horses:
        horseRaces = data.loc[data["horse_id"]==h]
        horseRaces = horseRaces.sort_values(by=["date"])
        
        raceSize = horseRaces.shape[0]
        raceI = horseRaces.index
        
        for i in range(1, raceSize):
            prevWins = wins[raceI[i-1]]
            wonLastRace = horseRaces.iloc[i-1,:]["won"]
            if wonLastRace==1:
                currentWin = prevWins + 1
            else:
                currentWin = prevWins
            wins[raceI[i]] = currentWin
            
    data["horse_wins"] = wins
    return data

# Add jockey number of wins
def addNumberOfWins(data):
    wins= np.zeros(len(data))
    
    jockeys= np.unique(data["jockey_id"])
    for h in jockeys:
        horseRaces = data.loc[data["jockey_id"]==h]
        horseRaces = horseRaces.sort_values(by=["date"])
        
        raceSize = horseRaces.shape[0]
        raceI = horseRaces.index
        
        for i in range(1, raceSize):
            prevWins = wins[raceI[i-1]]
            wonLastRace = horseRaces.iloc[i-1,:]["won"]
            if wonLastRace==1:
                currentWin = prevWins + 1
            else:
                currentWin = prevWins
            wins[raceI[i]] = currentWin
            
    data["jockey_wins"] = wins
    return data
```

```python
# main = winningsToDate(main)
# main = winningsPerRace(main)
main = addNumberOfWins(main)

main.to_pickle("../Data/main_1.df")
```
