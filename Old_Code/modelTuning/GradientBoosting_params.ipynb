```python
import sys
sys.path.append('../')
from helper import *
from combineData import *

from sklearn.ensemble import GradientBoostingClassifier

import matplotlib.pyplot as plt

runs = pd.read_pickle("../Data/main_1.df")

FEATURES = 
TARGET = "won"
```

```python
# Things to optimize:
# TREE STUFF - 
# min samples split, min samples leaf, max depth, max features,
# subsample
# Then lower learning rate and increase n_est
```

```python
# Baseline
bl_model = GradientBoostingClassifier()
meanScore, stdScore = crossVal(runs, FEATURES, TARGET, bl_model)
```

```python
# find min samples split
mss = [2, 3, 4, 5, 6, 7]
for split in mss:
    model = GradientBoostingClassifier(
        min_samples_split=split
    )
    print("Min sample split: {}".format(split))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# find loss 
vals = ["deviance", "exponential"]
for val in vals:
    model = GradientBoostingClassifier(
        min_samples_split=5,
        loss=val
    )
    print("Loss: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# min sample leaf
vals = [1,2,3,4,5,6,7]
for val in vals:
    model = GradientBoostingClassifier(
        min_samples_split=5,
        loss="deviance",
        min_samples_leaf=val,
    )
    print("Min samples leaf: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# max depth
vals = [1,2,3,4,5,6,7]
for val in vals:
    model = GradientBoostingClassifier(
        min_samples_split=5,
        loss="deviance",
        min_samples_leaf=3,
        max_depth=val
    )
    print("Max depth: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# subsample
vals = [0.5,0.6,0.7,0.8,0.9,1]
for val in vals:
    model = GradientBoostingClassifier(
        min_samples_split=5,
        loss="deviance",
        min_samples_leaf=3,
        max_depth=1,
        subsample=val,
    )
    print("Subsample: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```
