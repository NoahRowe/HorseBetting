```python
## PARAM TUNING STEPS

# Find the baseline

# max_depth

# min_samples_split

# min_samples_leaf

# max_features (?)
# Find best n_estimators
```

```python
# Baseline
bl_model = RandomForestClassifier()
meanScore, stdScore = crossVal(runs, FEATURES, TARGET, bl_model)
```

```python
# Min samples split
vals = [2, 3, 4, 5, 6, 7]
for val in vals:
    model = RandomForestClassifier(
        min_samples_split=val
    )
    print("Min sample split: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# Min samples leaf
vals = [1, 2, 3, 4, 5, 6]
for val in vals:
    model = RandomForestClassifier(
        min_samples_split=4,
        min_samples_leaf=val
    )
    print("Min sample split: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# Max depth - 7
vals = [None, 1, 3, 5, 7, 9, 20]
for val in vals:
    model = RandomForestClassifier(
        min_samples_split=4,
        min_samples_leaf=3,
        max_depth=val
    )
    print("Max depth: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```

```python
# Max features - 5
vals = [None, "auto", "sqrt", 10,20,30]
for val in vals:
    model = RandomForestClassifier(
        min_samples_split=4,
        min_samples_leaf=3,
        max_depth=5,
        max_features=val
        
    )
    print("Max features: {}".format(val))
    meanScore, stdScore = crossVal(runs, FEATURES, TARGET, model)
```
