```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier

runs = pd.read_pickle("Data/main_1.df")
runs = runs.iloc[int(len(runs)*.1):,:]
num_races = len(np.unique(runs["race_id"]))
# FEATURES = ["horse_no", "horse_age", "horse_rating", "declared_weight", "actual_weight", 
#             "win_odds", "draw", "race_size", "distance", "race_class"]
FEATURES = ["horse_no", "horse_age", "horse_rating", "declared_weight", "actual_weight", 
            "win_odds", "draw", #"race_size", "distance", "race_class", 
            'last_race_result','win_percent', 
            'avg_distance_time', 'normal_avg_distance_time',
            'going_type_record', 'actual_weight_scaled',
            'declared_weight_scaled', 'horse_race_count', "jockey_record",
            'trainer_record', 'horse_record', 'surface_record', 'place_odds',
            'weight_change', 'weight_change_over_time','weight_change_from_average', 'weight_change_increase',
            'venue_change','venue_record', 'days_since_last_race', 'new_horse',
            'best_odds', 'best_win_percent', 'best_distance_time', 'best_going_record', 
            'best_horse_record', 'best_jockey_record','best_trainer_record', 'highest_actual_weight', 
            'lowest_actual_weight', 'start_speed', 'rode_before',
#            'horse_type_Brown',
#            'horse_type_Colt', 'horse_type_Filly', 'horse_type_Gelding',
#            'horse_type_Grey', 'horse_type_Horse', 'horse_type_Mare',
#            'horse_type_Rig', 'horse_type_Roan', 'horse_country_ARG',
#            'horse_country_AUS', 'horse_country_BRZ', 'horse_country_CAN',
#            'horse_country_FR', 'horse_country_GB', 'horse_country_GER',
#            'horse_country_GR', 'horse_country_IRE', 'horse_country_ITY',
#            'horse_country_JPN', 'horse_country_NZ', 'horse_country_SAF',
#            'horse_country_SPA', 'horse_country_USA', 'horse_country_ZIM',
#           'going_type_fast', 'going_type_slow', 'going_type_wet'
           ]


TARGET = 'won'
#TARGET = 'placed'

X = runs[FEATURES]
y = runs[TARGET]

testPct = 0.2
trainIndex = int(num_races * (1-testPct))
max_race_id = np.unique(runs["race_id"])[trainIndex]
X_train = X.loc[runs["race_id"]<=max_race_id]
y_train = y.loc[runs["race_id"]<=max_race_id]
X_test = X.loc[runs["race_id"]>max_race_id]
y_test = y.loc[runs["race_id"]>max_race_id]
race_sizes_for_eval = [len(runs.loc[runs["race_id"]==race_id]["race_id"]) for race_id in np.unique(runs.loc[runs["race_id"]>max_race_id]["race_id"])]
```

```python
# MAKE A HEATMAP
heatMapData = runs[FEATURES + [TARGET]]
#get correlations of ach features in dataset
corrmat = heatMapData.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(heatMapData[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# Create models to show feature importance
# Random Forest Model
print("Training Random Forest model ...")
RFmodel = RandomForestClassifier()
RFmodel.fit(X_train, y_train)
# XGBoost model
print("Training XGBoost model ...")
XGBmodel = XGBClassifier()
XGBmodel.fit(X_train, y_train)
# LR model 
print("Training Logistic Regression model ...")
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)
# Gradient boosted tree
print("Training Gradient Boosting model ... ")
GBmodel = GradientBoostingClassifier()
GBmodel.fit(X_train, y_train)
```

```python
# Plot feature importances
N = len(FEATURES)
width = 0.15
ind = np.arange(N)

plt.figure(figsize=(10,8), dpi=100)
plt.bar(ind, RFmodel.feature_importances_, width, label="Random Forest")
plt.bar(ind+width, XGBmodel.feature_importances_, width, label="XGBoost")
plt.bar(ind+width*2, GBmodel.feature_importances_, width, label="Gradient Boosting")
LR_importance = abs(LRmodel.coef_)[0]
normal_LR = [val / float(sum(LR_importance)) for val in LR_importance]
plt.bar(ind+width*3, normal_LR, width, label="Logistic Regression")
plt.xticks(ind + width*3 / 2, FEATURES, rotation=90)
plt.legend()
```

```python
# Create our evaluate function
def winnerEval(model, x_test, y_test, race_sizes):
    # convert preds into an actual win choice
    winPreds = model.predict_proba(x_test)[:, 1]
    winCount = 0
    temp = 0 
    for i, s in enumerate(race_sizes):
        low_index = temp
        high_index = temp + s
        
        racePreds = winPreds[low_index:high_index]
        raceVals = y_test[low_index:high_index]
        
        if TARGET=='won':
            predWinner = np.argmax(racePreds, axis=0)
            actWinner = np.argmax(raceVals, axis=0)
        
            if predWinner == actWinner:
                winCount += 1
                
        elif TARGET=='placed':
            predPlacers = racePreds.argsort()[-3:]
            actPlacers = raceVals.argsort()[-3:]
            
            for val in actPlacers:
                if val in predPlacers:
                    winCount+=1
            
        temp += s
        

    if TARGET=='won':
        return winCount/float(len(race_sizes))
    else:
        return winCount / float(len(race_sizes)*3)

# DO ANOTHER FUNCTION THAT IS JUST FINDING THE BEST ODDS HORSE
def bestOddsEval(x_test, y_test, race_sizes):
    winCount = 0
    temp = 0 
    for i, s in enumerate(race_sizes):
        low_index = temp
        high_index = temp + s
        
        oddsPreds = x_test[low_index:high_index]
        raceVals = y_test[low_index:high_index]
        
        if TARGET=='won':
            predWinner = np.argmin(oddsPreds["win_odds"], axis=0)
            actWinner = np.argmax(raceVals, axis=0)

            if predWinner == actWinner:
                winCount += 1
            
        elif TARGET=='placed':
            predPlacers = oddsPreds["place_odds"].argsort()[:3].to_list()
            actPlacers = raceVals.argsort()[-3:].to_list()
            print(predPlacers, actPlacers)
            
            for val in actPlacers:
                if val in predPlacers:
                    winCount+=1
                    
        temp += s
        
    if TARGET=='won':
        return winCount/float(len(race_sizes))
    else:
        return winCount / float(len(race_sizes)*3)

def randEval(race_sizes):
    randCount = 0 
    for s in race_sizes:
        rand_a = np.random.randint(s)
        rand_b = np.random.randint(s)
        if rand_a==rand_b:
            randCount+=1
    return randCount/float(len(race_sizes))

randAcc = randEval(race_sizes_for_eval)
oddsAcc = bestOddsEval(X_test, y_test, race_sizes_for_eval)
        
print("Random Forest accuracy: {:.3f}".format(winnerEval(RFmodel, X_test, y_test, race_sizes_for_eval)))
print("Logistic Regression accuracy: {:.3f}".format(winnerEval(LRmodel, X_test, y_test, race_sizes_for_eval)))
print("Gradient Boosting accuracy: {:.3f}".format(winnerEval(GBmodel, X_test, y_test, race_sizes_for_eval)))
print("XGBoost accuracy: {:.3f}".format(winnerEval(XGBmodel, X_test, y_test, race_sizes_for_eval)))
print("Random guessing accuracy: {:.3f}".format(randAcc))
print("Betting best odds accuracy: {:.3f}".format(oddsAcc))
```

```python
# Perform univariate selection using a pipeline and k best
Ks = np.linspace(3, len(FEATURES), len(FEATURES)-2, dtype=int)
rfAccs = []
lrAccs = []
xgbAccs = []
gbAccs = []
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif, chi2
for k in Ks:
    print("K value: {}/{}".format(k, Ks[-1]))
    print("Training Random Forest model ... ")
    rf_pipe = Pipeline([('skb', SelectKBest(chi2, k = k)),
                        ('model', RandomForestClassifier())])
    rf_pipe.fit(X_train, y_train)
    print("Training Logisitc Regression model ... ")
    lr_pipe = Pipeline([('skb', SelectKBest(chi2, k = k)),
                        ('model', LogisticRegression())])
    lr_pipe.fit(X_train, y_train)
    print("Training XGBoost model ... ")
    xgb_pipe = Pipeline([('skb', SelectKBest(chi2, k = k)),
                        ('model', XGBClassifier())])
    xgb_pipe.fit(X_train, y_train)
    print("Training Gradient Boosting model ... ")
    gb_pipe = Pipeline([('skb', SelectKBest(chi2, k = k)),
                        ('model', GradientBoostingClassifier())])
    gb_pipe.fit(X_train, y_train)
    
    rfAcc = winnerEval(rf_pipe, X_test, y_test, race_sizes_for_eval)
    lrAcc = winnerEval(lr_pipe, X_test, y_test, race_sizes_for_eval)
    xgbAcc = winnerEval(xgb_pipe, X_test, y_test, race_sizes_for_eval)
    gbAcc = winnerEval(gb_pipe, X_test, y_test, race_sizes_for_eval)

    #print("Random Forest accuracy: {:.3f}".format(winnerEval(rf_pipe, X_test, y_test, race_sizes_for_eval)))
    #print("Logistic Regression accuracy: {:.3f}".format(winnerEval(lr_pipe, X_test, y_test, race_sizes_for_eval)))
    #print("Gradient Boosting accuracy: {:.3f}".format(winnerEval(gb_pipe, X_test, y_test, race_sizes_for_eval)))
    #print("XGBoost accuracy: {:.3f}".format(winnerEval(xgb_pipe, X_test, y_test, race_sizes_for_eval)))
    #print("Random guessing accuracy: {:.3f}".format(randAcc))
    #print("Betting best odds accuracy: {:.3f}".format(oddsAcc))
    
    rfAccs.append(rfAcc)
    lrAccs.append(lrAcc)
    xgbAccs.append(xgbAcc)
    gbAccs.append(gbAcc)
    
```

```python
# Plot the k value dependance
thresh = oddsAcc
N = len(Ks)
width = 0.15
ind = np.arange(N)

plt.figure(figsize=(10,8), dpi=100)
plt.bar(ind, rfAccs, width, label="Random Forest")
plt.bar(ind+width, xgbAccs, width, label="XGBoost")
plt.bar(ind+width*2, gbAccs, width, label="Gradient Boosting")
plt.bar(ind+width*3, lrAccs, width, label="Logistic Regression")
plt.plot([0, len(Ks)], [thresh,thresh], "k--")
plt.xticks(ind + width*3 / 2, Ks)
if TARGET=="won":
    plt.ylim(.20, .35)
else:
    plt.ylim(.4, 0.55)
plt.xlabel("K value")
plt.ylabel("Model accuracy")
plt.legend()
```
