```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from helper import *

main = pd.read_pickle("../clean_main.df")
```

```python
# Basically will use trainers record with a certain jockey (one that they call when they want a win or a scrap)

trainer_jockey_records = np.zeros(len(main))
trainers = np.unique(main["trainer_id"])
c = 0
for trainer in trainers:
    printProgress(c, len(trainers), jump = round(len(trainers)*0.05))
    c += 1
    trainer_races = main.loc[main.trainer_id == trainer]
    raceI = trainer_races.index
    
    for i in range(len(trainer_races)):
        if i == 0: # Trainers first race
            performance = trainer_races.iloc[0,:]["jockey_norm_record"]
        else:
            jockey = trainer_races.iloc[i,:]["jockey_id"]
            past_races = trainer_races.iloc[:i,:]
            past_jockey_races = past_races.loc[past_races.jockey_id == jockey]
            
            if len(past_jockey_races) == 0:
                performance = trainer_races.iloc[i,:]["jockey_norm_record"]
            else:
                performance = past_jockey_races["norm_result"].mean()
                
        trainer_jockey_records[raceI[i]] = performance
        
main["trainer_jockey_record"] = trainer_jockey_records
```

```python
main.to_pickle("../clean_main.df")
```
