```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main = pd.read_pickle("../Data/main_1.df")
badRaceList = []
for i in range(len(main)):
    run = main.iloc[i,:]
    if run["lengths_behind"]>40:
        race_id = run["race_id"]
        badRaceList.append(race_id)
indexNames = main[main['race_id'].isin(np.unique(badRaceList))].index
main.drop(indexNames, inplace=True)
```

```python
indexNames = main[main['race_id'].isin(np.unique(badRaceList))].index
main.drop(indexNames, inplace=True)
```

```python
import seaborn as sns
plt.figure(figsize=(10,8), dpi=100)
sns.violinplot(data=main, x="lengths_behind")
plt.xlim(0, 100)
```
